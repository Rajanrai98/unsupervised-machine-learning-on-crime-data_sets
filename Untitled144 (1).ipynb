```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3-Q1
```

```python
IQR
```

```python
upper_bound = Q3+1.5*IQR
lower_bound = Q1-1.5*IQR
```

```python
upper_bound , lower_bound
```

```python
data = data[~((data>upper_bound)|(data<lower_bound))]
```

```python
data
```

```python
sn.boxplot(data)
```

```python
data = data.drop(['Unnamed: 0'] , axis = 1)
```

```python
data
```

```python
data.isnull().sum()
```

```python
mean = data['Rape'].mean()
```

```python
data = data.fillna(mean)
```

```python
data
```

```python
def norm_func(i):
    x = (i - i.min()/i.max()-i.min())
    return(x)
```

```python
X = norm_func(data)
```

```python
X
```

```python
x = np.array(X)
```

```python
x
```

```python
hc_cluster = AgglomerativeClustering(n_clusters = 3 , linkage = 'ward')
hc_cluster.fit(x)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
z =linkage(x  , method = 'ward')
plt.figure(figsize = (12,8))
dendrogram(z)
plt.title('dendrogram')
plt.xlabel('data_points')
plt.ylabel('distance')
plt.show
```

## KMEANS Clustering

```python
k_cluster = KMeans(n_clusters = 3 , random_state = 42)
```

```python
k_cluster.fit(x)
```

```python
wcss = []
for k in range(1,12):
    kmean = KMeans(n_clusters = k , random_state = 42)
    kmean.fit(x)
    wcss.append(kmean.inertia_)
```

```python
number_cluster = range(1,12)
plt.plot(number_cluster , wcss)
plt.title('Elbow method')
plt.xlabel('number of cluster')
plt.ylabel('with in cluster sum of square')
```
